#**Task**
Write a program that asks for the user's name and age, and then displays a message with that information.
1. Use input() to ask the user for their name and age.
2. Store the user's name and age in variables.
3. Use print() to display a greeting message with the user's name and age.

```python
# Ask the user for their name
name = input("Enter your name: ")

# Ask the user for their age
age = input("Enter your age: ")

# Display greeting message
print("Hello", name + "! You are", age, "years old.")
```
