```python
import pandas as pd
df=pd.read_csv("Dataset_Grades.csv")
```

```python

# Remove Seat No column
df = df.drop("Seat No.", axis=1)
```

```python

df.head(5)
```

```python
df.isnull().sum()
```

```python
df = df.fillna(df.mode().iloc[0])

df.isnull().sum()
```

```python
# df = df.fillna(method="bfill")
```

```python
# A > A- > B+ > B > B- > C+ > C > C- > D+ > D
# ordinal encding
grade_map = {
'A': 4.0,
'A-': 3.7,
'B+': 3.3,
'B': 3.0,
'B-': 2.7,
'C+': 2.3,
'C': 2.0,
'C-': 1.7,
'D+': 1.3,
'D': 1.0,
'F': 0
}
```

```python
for col in df.columns[:-1]:   # exclude CGPA
    df[col] = df[col].map(grade_map)
```
