```python
# part a 
import numpy as np

# Function
def J(t1, t2, t3):
    return t1**2 + t2**2 + t3**2

# Partial derivatives
def dJ_dt1(t1, t2, t3):
    return 2*t1

def dJ_dt2(t1, t2, t3):
    return 2*t2

def dJ_dt3(t1, t2, t3):
    return 2*t3
```

```python
# Hyperparameters
alpha = 0.02
iterations = 5000

# Initial values (set appropriately, e.g., far from minimum)
t1, t2, t3 = 10.0, 10.0, 10.0

print(f"{'Iter':<10} {'t1':<12} {'t2':<12} {'t3':<12} {'Cost J':<12}")
print("-" * 60)

for i in range(iterations):
    # Compute gradients
    grad_t1 = dJ_dt1(t1, t2, t3)
    grad_t2 = dJ_dt2(t1, t2, t3)
    grad_t3 = dJ_dt3(t1, t2, t3)

    # Simultaneous update
    t1 = t1 - alpha * grad_t1
    t2 = t2 - alpha * grad_t2
    t3 = t3 - alpha * grad_t3

    cost = J(t1, t2, t3)

    if i < 5 or i == iterations - 1:
        print(f"{i+1:<10} {t1:<12.6f} {t2:<12.6f} {t3:<12.6f} {cost:<12.6f}")

print(f"\nFinal -> t1={t1:.6f}, t2={t2:.6f}, t3={t3:.6f}, Cost={J(t1,t2,t3):.6f}")
```

```python
# Part c: Manual Trace for First 3 Iterations
# Starting values: t1 = t2 = t3 = 10, α = 0.02
# Iteration 1:

# Gradients: ∂J/∂t1 = 2(10)=20, ∂J/∂t2 = 20, ∂J/∂t3 = 20
# t1 = 10 − 0.02×20 = 9.6
# t2 = 10 − 0.02×20 = 9.6
# t3 = 10 − 0.02×20 = 9.6
# Cost = 9.6² + 9.6² + 9.6² = 276.48
# Iteration 2:

# Gradients: 2(9.6) = 19.2 each
# t1 = t2 = t3 = 9.6 − 0.02×19.2 = 9.216
# Cost = 3 × 9.216² = 254.804

# Iteration 3:

# Gradients: 2(9.216) = 18.432 each
# t1 = t2 = t3 = 9.216 − 0.02×18.432 = 8.847
# Cost = 3 × 8.847² = 234.820
```

```python
# Part d: Has it converged to the global minimum?
# Yes, with 5000 iterations and α = 0.02, the algorithm converges to the global minimum.

# The global minimum of f(t1, t2, t3) = t1² + t2² + t3² is 0, achieved at (t1, t2, t3) = (0, 0, 0)
# This function is strictly convex (bowl-shaped), so gradient descent is guaranteed to find the global minimum
# The update rule shrinks each parameter by a factor of (1 − 2α) = 0.96 per iteration → exponentially decays to 0
# After 5000 iterations: t ≈ 10 × (0.96)^5000 ≈ 0 ✅


# No fix needed — unlike functions with local minima, this quadratic function has a single global minimum and gradient descent reliably finds it.
```
