```python
# Q1. Import Required Libraries
# Import all the necessary Python libraries for data handling and linear regression.

# Example libraries: pandas, numpy, matplotlib, sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Q2. Load the Dataset
# Load the dataset into a DataFrame using pandas.read_csv()
# Display the first few rows using print

df = pd.read_csv("score.csv")
print(df.head())
```

```python
# Q3. Prepare the Data
# Separate the features (Hours) and target (Scores)

X = df[['Hours']]
y = df['Scores']
```

```python
# Q4. Train the Simple Linear Regression Model
# Import LinearRegression from sklearn.linear_model.
# Fit the model on all the data using .fit().

# Create the Linear Regression model
model = LinearRegression()

# Fit the model on all the data
model.fit(X, y)

# Optional: print the slope and intercept
print("Coefficient (Slope):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
# Make Predictions
# Use the trained model to predict scores using .predict(X).
# Display both actual and predicted scores side by side.
# Q5. Make Predictions

# Predict scores using the trained model
y_pred = model.predict(X)

# Display actual and predicted scores side by side
results = pd.DataFrame({
    "Actual": y,
    "Predicted": y_pred
})

print(results)
```

```python
# Predict a custom value without warnings

# Create a DataFrame with the same column name as training data
hours_df = pd.DataFrame({'Hours': [9.25]})

# Make prediction
predicted_score = model.predict(hours_df)

print(f"Predicted score for a student studying 9.25 hours: {predicted_score[0]:.2f}")
```
