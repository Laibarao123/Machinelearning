```python
# STEP 1 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# STEP 2
df=pd.read_csv(r"C:\Users\dell\Downloads\scores.csv")
df.head()
```

```python
# step 3
X=df[['Hours']]
y=df['Scores']
```

```python
model = LinearRegression()
model.fit(X, y)
print(f"Slope: {model.coef_[0]:.2f}")
print(f"Intercept: {model.intercept_:.2f}")
```

```python
y_pred = model.predict(X)
results = pd.DataFrame({'Actual Scores': y, 'Predicted Scores': y_pred.round(2)})
print(results)
```

```python
predicted = model.predict([[9.25]])
print(f"Predicted score for 9.25 hours: {predicted[0]:.2f}")
```
