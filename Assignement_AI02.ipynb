##**Part 3: Type Conversion**

####**Write a program that:**

- Takes user input for age
- Converts it to integer
- Adds 5 to it
- Prints the new age

⚠️ Remember: input() returns string.

```python
# Integer variable for age
age = 10

# Float variable for height
height = 5.1

# String variable for city
city = "Karachi"

# Boolean variable
is_student = True

# Print variables with their types
print("Age:", age, type(age))
print("Height:", height, type(height))
print("City:", city, type(city))
print("Is Student:", is_student, type(is_student))
```

```python
# Take user input
age = input("Enter your age: ")

# Convert to integer and add 5
age_plus_five = int(age) + 5

# Print result
print(f"In 5 years, you will be {age_plus_five} years old.")
```
