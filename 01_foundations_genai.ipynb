
### 2.3 What are Tokens?

LLMs don't read words like humans do. They read "tokens".

- **Token:** A chunk of text. It can be a word (`apple`), part of a word (`ing`), or a character (`@`).
- **Rule of Thumb:** 100 tokens is roughly 60-80 English words.

We can check how many tokens a text uses:


````python
text = """
Artificial intelligence is transforming the world rapidly.
Tokenization does not always match word boundaries.
"""

# ---- Count Words ----
word_count = len(text.split())

# ---- Count Tokens ----
token_response = model.count_tokens(text)
token_count = token_response.total_tokens

# ---- Print Results ----
print("Text:")
print(text)
print("-" * 40)
print(f"Word count: {word_count}")
print(f"Token count: {token_count}")
````


### 2.4 Context Window

The **Context Window** is the maximum amount of information (tokens) the model can hold in its "short-term memory" at once.

- If your conversation exceeds this limit, the model "forgets" the earliest parts.
- **Gemini 1.5 Flash** has a massive context window of **1 Million tokens** (enough for several books!).



## 3. Introduction to Prompting

The quality of the output depends on the quality of your input (the **Prompt**).

**Components of a Good Prompt:**
1.  **Instruction:** What do you want done? (e.g., "Summarize this")
2.  **Context:** Background info. (e.g., "You are a teacher")
3.  **Input Data:** The content to process.
4.  **Output Format:** How you want the answer. (e.g., "Bullet points")


````python
bad_example = "Create a email to sell water bottles"

good_example = """
You are a expert marketer for company called DataCrumbs, your job is to create the best sales email that i can sent to customers
so they can buy my water bottles. The product is prices at 15 dollars and it has insulation too
Be polite and use good hooks, and the maximum number of words should be 100
"""

print("---Bad Prompt---")
response = model.generate_content(bad_example)
# print(response.text)

print("---Good Prompt---")
response = model.generate_content(good_example)
print(response.text)
````


### 3.1 Zero-Shot Prompting

**Zero-shot** means asking the model to do something without giving it any examples. You just rely on its general knowledge.

- **Example:** "Translate 'Hello' to Spanish."



### 3.2 Few-Shot Prompting

**Few-shot** means giving the model a few examples (shots) of what you want, so it understands the pattern.

This is very powerful for specific formats or tricky logic.
