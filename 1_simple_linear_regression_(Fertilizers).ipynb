# Simple Linear Regression

## Importing the libraries

```
# import libraries
import pandas as pd # Used for dealing with excel data
```

## Importing the dataset

```
# Load the dataset
data = pd.read_csv("Fertilizers.csv")

# Display the dataset
print(data)

# Define features (X) and target (Y)
X = data[["Nitrogen_Fertilizer_kg_per_acre"]] #Its a X variable [independent variable]
y = data["Potato_Yield_tonnes_per_acre"] # Its a y variable [dependent variable]
```

```
print(X)
```

```
print(y)
```

## Splitting the dataset into the Training set and Test set

```
# Split the dataset into training and test sets (80% used for training, 20% used for testing)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```
print(X_train)
```

```
print(X_test)
```

## Training the Simple Linear Regression model on the Training set

```
# Create and train the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

## Predicting the Test set results

```
# Predict on test data
y_pred = model.predict(X_test)
print(y_pred)
```

## Visualising the Training set results

```
import matplotlib.pyplot as plt

# Scatter plot of actual data
plt.scatter(X, y, color='blue', label='Actual Data')

# Optional: plot regression line
plt.plot(X, model.predict(X), color='red', label='Regression Line')

# Labels and title
plt.xlabel("Nitrogen Fertilizer (kg per acre)")
plt.ylabel("Potato Yield (tonnes per acre)")
plt.title("Nitrogen Fertilizer vs Potato Yield")
plt.legend()

# Show plot
plt.show()
```
