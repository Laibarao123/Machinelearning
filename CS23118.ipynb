```python
df.isnull().sum()
```

```python
# Strip column names first
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(r'\s+', ' ', regex=True)
```

```python

# Alcohol, Polio, Diphtheria: fill by Country median
for col in ['Alcohol', 'Polio', 'Diphtheria']:
    df[col] = df.groupby('Country')[col].transform(lambda x: x.fillna(x.median()))

# Total expenditure, GDP, Income composition of resources, Schooling: Country-wise median
for col in ['Total expenditure', 'GDP', 'Income composition of resources', 'Schooling']:
    df[col] = df.groupby('Country')[col].transform(lambda x: x.fillna(x.median()))
```

```python
# For any leftover missing values
df.fillna(df.median(numeric_only=True), inplace=True)
```

# Feature Engineering

# Create new features

```python
# a) Mortality Ratio

# Combine Adult Mortality and Infant deaths into a single measure:
df['Mortality Ratio'] = (df['Adult Mortality'] + df['infant deaths']) / df['Population']
# Justification:

# Normalizes mortality by population, giving a better comparison across countries of different sizes.

# Helps in understanding which countries have disproportionately high mortality rates.
```

```python
# b) Vaccination Coverage

# Combine Polio, Diphtheria, Measles, Hepatitis B into an average vaccination rate:
df['Vaccination Coverage'] = df[['Polio', 'Diphtheria', 'Measles', 'Hepatitis B']].mean(axis=1)
# Justification:

# Single feature summarizing immunization performance

# Useful for predicting Life expectancy or public health metrics.
```

```python
# c) Health Expenditure per GDP

# If you have Total expenditure and GDP:
df['Health Expenditure % GDP'] = df['Total expenditure'] / df['GDP']
# Justification:

# Standardizes spending relative to country wealth

# Useful for analyzing the impact of expenditure on health outcomes.
```

```python
# ) BMI Category (Categorical)

# Transform BMI into categories:
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 25:
        return 'Normal'
    elif bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

df['BMI Category'] = df['BMI'].apply(bmi_category)
# Justification:

# Converts continuous BMI to interpretable categories

# Useful for classification models or demographic analysis.
```

```python
# e) Child Mortality Rate

# Normalize under-five deaths by Population:
df['Under5 Mortality Rate'] = df['under-five deaths'] / df['Population']
# Justification:

# Easier comparison across countries

# Better predictor for Life expectancy.
```

# Transform existing features

# a) Scaling numeric features

```python
from sklearn.preprocessing import StandardScaler

num_cols = ['Life expectancy', 'Adult Mortality', 'BMI', 'Alcohol', 'GDP', 'Income composition of resources', 'Schooling']
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
# Justification:

# Puts features on same scale → avoids models being dominated by large-value features like GDP.
```

# b) Encoding categorical variables

```python
# Status → Developed/Developing → binary or one-hot encoding

# BMI Category → one-hot encoding
df = pd.get_dummies(df, columns=['Status', 'BMI Category'], drop_first=True)
# Justification:

# Converts categorical variables into numeric form → required for most ML algorithms

# Preserves information while avoiding arbitrary ordinal assignment.
```

# Histogram for Numeric Distributions

```python
# Interpretation

# Shape of Distribution:

# If the histogram is right-skewed → most countries have high life expectancy, few countries have very low life expectancy

# If bell-shaped / normal → life expectancy is fairly evenly spread around a central value

# Central Tendency:

# Peak of the histogram → most common life expectancy range

# You can roughly estimate the median/mean from the peak

# Spread / Variability:

# Wide spread → big differences in life expectancy between countries

# Narrow spread → countries have similar life expectancy

# Outliers:

# Countries at extreme low values → likely underdeveloped or conflict-affected countries

# Countries at extreme high values → likely developed countries
```

# Box Plot: Life Expectancy by BMI Category
