## **Plot the decision tree**

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# Load dataset
df = pd.read_csv('data.csv')

# Convert columns to numeric (IMPORTANT FIX)
df['Level'] = pd.to_numeric(df['Level'])
df['Salary'] = pd.to_numeric(df['Salary'])

# Features and target
X = df[['Level']].values
y = df['Salary'].values

# Train Decision Tree Regressor
model = DecisionTreeRegressor(random_state=0)
model.fit(X, y)

# ----- Decision Tree Regression Plot -----
X_grid = np.arange(X.min(), X.max(), 0.01)
X_grid = X_grid.reshape(-1, 1)

plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='red', label='Actual data')
plt.plot(X_grid, model.predict(X_grid), color='blue', label='Decision Tree Prediction')
plt.title('Decision Tree Regression: Salary vs Level')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.legend()
plt.show()

# ----- Predict salary for Level 7 -----
predicted_salary = model.predict([[7]])
print(f"Predicted salary for level 7: ${predicted_salary[0]:,.0f}")
```

## **Visualising the results**

```
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
from sklearn import tree

# Load the dataset from the CSV file
df = pd.read_csv('data.csv')

# Inspect the first few rows of the data
print(df.head())

# Define features (X) and target (y)
X = df[['Level']]  # Feature: Level
y = df['Salary']  # Target: Salary

# Create and fit the decision tree model
model = DecisionTreeRegressor()
model.fit(X, y)

# Plot the decision tree
plt.figure(figsize=(12,8))
tree.plot_tree(model, feature_names=['Level'], filled=True)
plt.show()

# Predictions can be made using model.predict() on new data
```
